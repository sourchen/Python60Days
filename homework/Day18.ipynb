```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# sin波 從0~180度
x = np.arange(0,180)
y = np.sin(x * np.pi / 180.0)
plt.plot(x,y)
plt.show() # 秀圖
```

```python
#完整的 sin
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
plt.plot(x, y_sin)
plt.show()
```
